```python
import pandas
import json
import os
import gc

def convert_to_parquet(mode: str = 'niid', dataset: str = 'train'):
  directory = os.path.join(f'./data/{mode}/{dataset}')
  files = os.listdir(directory)

  def create_json_from_json(path):
    with open(os.path.join(directory, path), 'r') as f:
      data = json.load(f)
  
    exploded = pandas.DataFrame(data['user_data']).T.explode(list('xy')) 

    del data
    gc.collect()

    os.mkdir(f'output_{mode}_{dataset}')
    os.mkdir(f'output')

    features = pandas.DataFrame(exploded['x'].values.tolist(), index = exploded.index)
    features.columns = [ f'x{i + 1}' for i in range(784) ]

    flattened = pandas.concat([ exploded.drop(columns = 'x'), features ], axis = 1)
    flattened.to_parquet(f"output_{mode}_{dataset}/{name.removesuffix('json')}.parquet", index_label = 'user')

    del exploded
    del flattened
    del features
    gc.collect()

  # all json to all parquet
  for i, name in enumerate(files):
    create_csv_from_json(name)
    print(f'...[{i + 1}/{len(files)}]')

  gc.collect()
  
  directory = os.path.join(f'output_{mode}_{dataset}')
  files = os.listdir(directory)
  frames = []

  # all parquet to one unique parquet
  for i, name in enumerate(files):
    frames.append(pandas.read_parquet(os.path.join(directory, name)))

  pandas.concat(frames, axis = 0).to_parquet(f'output/{mode}/{dataset}.parquet')

  gc.collect()
```
